```python
import types
import sys;
from pathlib import Path
from datasets import build_dataset
import datasets.transforms as T
import datasets.coco as coco
import os

# -------------------------- 配置参数 --------------------------
# 替换为你的实际数据集路径（原命令中的 $DATA 环境变量）
data_path_str = os.path.expanduser("~/autodl-tmp/DrivingSDK/model_examples/Deformable-DETR/data/waterscenes-coco")
DATA_PATH = Path(data_path_str)
DATA = DATA_PATH
ANN = f"{DATA}\\instances_train.json"
# 检查的样本数量（保持原命令的数值）
CHECK_NUM_1 = 500  # 第一部分检查的最大样本数
CHECK_NUM_2 = 200  # 第二部分检查的最大样本数

def check_empty_by_build_dataset():
    """
    第一部分：通过 build_dataset 检查训练集空 boxes 比例
    """
    # 构造命令行中的 args 参数（与原命令保持一致）
    args = types.SimpleNamespace(
        coco_path=DATA,
        dataset_file='coco',
        masks=False,
        cache_mode=False,
        tir_mean=0.5,
        tir_std=0.5,
        tir_dropout=0.3,
        radar_enabled=True,
        radar_root='',
        radar_calib='auto',
        radar_range_max=200.0,
        radar_doppler_max=20.0,
        radar_elev_max=10.0,
        radar_power_max=20.0
    )
    
    # 构建训练集并检查空 boxes
    print("=== 第一部分：build_dataset 检查空标注 ===")
    ds = build_dataset('train', args)
    n = min(len(ds), CHECK_NUM_1)
    empty = 0
    for i in range(n):
        # 遍历样本，统计无 boxes 的数量
        _, target = ds[i]
        if target.get('boxes') is None or len(target['boxes']) == 0:
            empty += 1
    # 输出结果
    print(f"检查样本数: {n}, 空标注数: {empty}, 空标注比例: {empty/n:.4f}")

def count_empty(transforms, root, ann_file, check_num):
    """
    辅助函数：统计指定 transforms 下的空标注数量
    Args:
        transforms: 数据增强组合
        root: 数据集根路径
        ann_file: 标注文件路径
        check_num: 检查的最大样本数
    Returns:
        empty: 空标注数量
        n: 实际检查的样本数
    """
    ds = coco.CocoDetection(
        image_root=Path(root) / 'image',  # 原命令中的 root/'image'
        ann_file=ann_file,
        transforms=transforms,
        return_masks=False,
        cache_mode=False,
        local_rank=0,
        local_size=1,
        tir_root=None,
        radar_root=None,
        radar_enabled=True,
        radar_calib=None,
        radar_norm=None
    )
    n = min(len(ds), check_num)
    empty = 0
    for i in range(n):
        _, target = ds[i]
        if target.get('boxes') is None or len(target['boxes']) == 0:
            empty += 1
    return empty, n

def check_empty_by_transforms():
    """
    第二部分：对比带/不带 RandomSizeCrop 的空标注比例
    """
    print("\n=== 第二部分：不同数据增强下的空标注对比 ===")
    # 定义带 RandomSizeCrop 的完整数据增强
    full_transforms = T.Compose([
        T.RandomHorizontalFlip(),
        T.RandomSelect(
            None,
            T.Compose([
                T.RandomResize([400, 500, 600]),
                T.RandomSizeCrop(384, 600)
            ])
        ),
        T.RandomResize([800], max_size=1333),
        T.ModalityDropout(p_tir=0.0, p_radar=0.0),
        T.ToTensor(),
        T.Normalize(
            [0.485, 0.456, 0.406, 0.5, 0.0],
            [0.229, 0.224, 0.225, 0.5, 1.0]
        )
    ])
    
    # 定义不带 RandomSizeCrop 的数据增强
    no_crop_transforms = T.Compose([
        T.RandomResize([800], max_size=1333),
        T.ToTensor(),
        T.Normalize(
            [0.485, 0.456, 0.406, 0.5, 0.0],
            [0.229, 0.224, 0.225, 0.5, 1.0]
        )
    ])
    
    # 统计两种增强策略的空标注比例
    e1, n1 = count_empty(full_transforms, DATA, ANN, CHECK_NUM_2)
    e2, n2 = count_empty(no_crop_transforms, DATA, ANN, CHECK_NUM_2)
    
    # 输出结果
    print(f"带 RandomSizeCrop 空标注比例: {e1/n1:.4f} (检查 {n1} 样本)")
    print(f"无 RandomSizeCrop 空标注比例: {e2/n2:.4f} (检查 {n2} 样本)")

if __name__ == "__main__":
    # 执行主逻辑
    check_empty_by_build_dataset()
    check_empty_by_transforms()
```
